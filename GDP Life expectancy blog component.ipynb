# Does wealth = health? How GDP and Life expectancy trend together

```python
# load key packages and load data

from matplotlib import pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

data = pd.read_csv('all_data.csv')
```

#### Countries are often ranked by the wealth they generate using a crude measure known as gross domestic product (GDP), but what is GDP and does it have any tangible meaning in terms of the life experience of citizens of individual nations?

#### GDP is the value of all goods and services produced by a country over a given time period, often a calendar year. For many years it has been used to determine how far countries have developed economically, with an underlying assumption that higher GDP relates to more advanced economies and a better quality of life for people living in a country. 

#### In recent years, these assumptions have been challenged in light of pushback against the doctrine that continued economic growth is both desirable and achievable. As the human population has grown dramatically over the past 100 years, pressure on the Earth's resources and ecosystems has hit a critical point and the world economy will need to adapt if people are to continue to flourish. 

#### In addition, a single measure per country fails to detect large differences in the share of a nation's wealth that its citizens may experience. With growing wealth divides a common feature around the world, a large GDP may be reported where total revenues are high but profits are accumulated in the bank accounts of a small minority of people. 

#### With these issues in mind, I sought to investigate whether GDP is linked to life expectancy based on the data of six different countries collected between 2000 and 2015. It should be noted that these analyses were based on total GDP, not GDP per capita, which takes the size of a country's population into account.   

```python
country_data = data.groupby('Country')

chile = country_data.get_group('Chile')
china = country_data.get_group('China')
germany = country_data.get_group('Germany')
mexico = country_data.get_group('Mexico')
usa = country_data.get_group('United States of America')
zimbabwe = country_data.get_group('Zimbabwe')

plt.figure(figsize=(12, 8))

plt.subplot(2,3,1)
sns.scatterplot(data=chile, x='GDP', y='Life expectancy at birth (years)', color='blue')
plt.title('Chile')

plt.subplot(2,3,2)
sns.scatterplot(data=china, x='GDP', y='Life expectancy at birth (years)', color='orange')
plt.title('China')

plt.subplot(2,3,3)
sns.scatterplot(data=germany, x='GDP', y='Life expectancy at birth (years)', color='green')
plt.title('Germany')

plt.subplot(2,3,4)
sns.scatterplot(data=mexico, x='GDP', y='Life expectancy at birth (years)', color='red')
plt.title('Mexico')

plt.subplot(2,3,5)
sns.scatterplot(data=usa, x='GDP', y='Life expectancy at birth (years)', color='purple')
plt.title('USA')

plt.subplot(2,3,6)
sns.scatterplot(data=zimbabwe, x='GDP', y='Life expectancy at birth (years)', color='brown')
plt.title('Zimbabwe')

g1 = plt.subplot(2,3,1)
g2 = plt.subplot(2,3,2)
g3 = plt.subplot(2,3,3)
g4 = plt.subplot(2,3,4)
g5 = plt.subplot(2,3,5)
g6 = plt.subplot(2,3,6)

g1.set(xlabel=None)
g2.set(ylabel=None)
g2.set(xlabel=None)
g3.set(xlabel=None)
g3.set(ylabel=None)
g5.set(ylabel=None)
g6.set(ylabel=None)

plt.show()
```

#### As you can see, if we consider how life expectancy at birth changes with increasing GDP, the general trend is very much that higher GDP is associated with higher life expectancy.   

#### But it is also worth looking closely at the axes of these individual graphs and you will see that both life expectancy and GDP vary considerably between nations.  

```python
sns.scatterplot(x="GDP", y="Life expectancy at birth (years)", hue="Country", data=data)
plt.title('Impact of GDP on life expectancy in 6 different countries')
plt.show()
```

#### When all the data are combined into a single graph, the differences between the different countries become more apparent.  

#### Whilst it is true that a positive association between GDP and life expectancy is evident in all 6 countries, the slope of these associations is generally steeper at lower GDPs. This indicates that as a country's GDP increases, whilst life expectancy continues to improve, the gains decrease.  

#### These diminishing returns suggest that there may be a peak in economic productivity beyond which there will be no further life expectancy gains to be had by simply increasing a country's wealth.  

#### So does health = wealth? To a degree, the answer is yes in so far as considering two broad measures: GDP and average life expectancy from birth suggests concurrent gains for both.  

#### There could be some obvious explanations for this, if people live longer, they will be able to work for longer, increasing a country's workforce and helping productivity. From the other perspective, if a country is wealthy it will have more money to invest in its healthcare system leading to improvements in life expectancy.  

#### These are just a couple of ways in which it can be imagined that wealth and health are inter-dependent. However, it is important to note that even within the wealthiest of nations there are great differences between the wealth and health experiences of different groups of citizens. Therefore, the use of such broad measures fails to pick up on the complete story of the relationship between a country's wealth and the health of its people.
